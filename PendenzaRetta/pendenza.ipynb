```python
import numpy as np
import matplotlib.pyplot as plt
```

# Definizione della funzione


```python
def f(x):
    return (x**3-x**2-2*x-3)
```

Verfica

```python
print(f(0.5))
```

Definiamo un vettore che contenga i valri x da -10 fino a 10 ogni 0.01, sono 201 valori

```python
x =np.linspace(-10,10,201)
print(x)
```

```python
y = f(x)
```

Scegliamo un valore per $\Delta x$ 

```python
dx=0.01
```

```python
m=(f(x+dx)-f(x))/dx
print(m)
```

Rappresentiamo graficamente $f(x)$ e la sua pendenza $m(x)$

```python
griglia=np.linspace(-100,100,30)

plt.figure(figsize=(7,7))
plt.plot(x,y)
plt.plot(x,m)
plt.axis([-10,10,-10,10])
plt.gca().set_aspect("equal")
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid()
plt.show()
```

Esegiamo una regeressione polinomiale su $m(x)$

```python
np.polyfit(x,m,2)
```
